## Image to text API

```python
import requests
from api_keys import covid_api
from pprint import pprint
```

```python
url = 'https://api.api-ninjas.com/v1/imagetotext'
image_file_descriptor = open('text.png', 'rb')
files = {'image': image_file_descriptor}
params = {
    'X-Api-Key': covid_api
}
r = requests.post(url, files=files, params = params)
print(r.json())
```

```python
r = r.json()
for i in range(len(r)):
    print (r[i]['text'])
```
