```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fp = 'labour_data.csv'
data = pd.read_csv(fp)
data.dropna(axis = 1, thresh = 5)
data['Date'] = pd.to_datetime(data['Date'])
data = data.set_index('Date')
for col in data.columns:
    data[col] = data[col].replace(',', '', regex=True).astype(float)
```
